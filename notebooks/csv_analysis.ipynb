# Análisis de Sentimientos en Textos con Python
Este proyecto tiene como objetivo analizar sentimientos en textos a partir de un archivo CSV. Utiliza modelos de análisis de sentimientos, realiza visualizaciones y crea una nube de palabras para explorar los datos de manera visual. El análisis está basado en los datos de texto con la correspondiente etiqueta de sentimiento y un valor de confianza asociado.

## Importar librerías necesarias

pandas: para manipulación de datos.
matplotlib: para visualización de datos.
wordcloud: para crear nubes de palabras.



```python
# Importamos las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## Cargar el archivo CSV

Cargamos el archivo CSV que contiene los resultados del análisis de sentimientos. Este archivo debe contener al menos tres columnas: "Texto", "Sentimiento" y "Confianza". La columna "Texto" contiene el texto analizado, la columna "Sentimiento" contiene la etiqueta de sentimiento (por ejemplo, positivo, negativo o neutral) y la columna "Confianza" contiene un valor numérico que indica la confianza del modelo en su predicción.


```python
file_path = "./../src/sentiment_analysis_results.csv"
data = pd.read_csv(file_path)
```

## Analizar y visualizar los sentimientos

### 1. Verificar la estructura del DataFrame

Verificamos la estructura del DataFrame para asegurarnos de que los datos se han cargado correctamente. Esto nos permitirá entender mejor la información que estamos analizando.









```python
# 3. Analizar distribución de los sentimientos
sentiment_counts = data['Sentimiento'].value_counts()

# Visualizar la distribución de los sentimientos
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color='skyblue')
plt.title('Distribución de los Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# 4. Analizar la confianza en los sentimientos
plt.figure(figsize=(8, 6))
plt.hist(data['Confianza'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribución de la Confianza en los Sentimientos')
plt.xlabel('Confianza')
plt.ylabel('Frecuencia')
plt.show()
```

# Crear una nube de palabras

Una nube de palabras es una representación visual de la frecuencia de palabras en un texto. Las palabras más frecuentes aparecen en un tamaño más grande, lo que permite identificar rápidamente los términos más relevantes.

```python
# 5. Crear una nube de palabras con los textos
text = ' '.join(data['Texto'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras de los Textos')
plt.show()
```

# Análisis de tipos de sentimientos





```python
sentiment_analysis = data.groupby('Sentimiento')['Confianza'].mean()

# Visualizar el análisis de sentimiento
plt.figure(figsize=(8, 6))
sentiment_analysis.plot(kind='bar', color='lightgreen')
plt.title('Confianza Promedio por Sentimiento')
plt.xlabel('Sentimiento')
plt.ylabel('Confianza Promedio')
plt.xticks(rotation=45)
plt.show()
```
